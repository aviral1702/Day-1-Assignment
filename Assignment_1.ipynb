```python
import numpy as np
import pandas as pd
import seaborn as sns
 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
 
print('Matplotlib version',mpl.__version__)
```

```python
x=np.arange(0,10)
y=x*x
```

```python
plt.plot(x,y,'ro--',linewidth=1,markersize=5)
plt.title("Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
days=[1,2,3,4,5,6,7]
sales_1=[160,150,140,145,175,165,180]
sales_2=[70,90,160,150,140,145,175]
```
